```python
df_name=rating.merge(books,on='ISBN')
df_name.head()
```

```python
num_rating_df=df_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_rating'},inplace=True)
num_rating_df.head()
```

```python
df_name['Book-Rating'] = pd.to_numeric(df_name['Book-Rating'], errors='coerce')
avg_rating_df = (
    df_name.groupby('Book-Title', as_index=False)['Book-Rating']
    .mean()
    .rename(columns={'Book-Rating': 'avg_rating'})
)

avg_rating_df.head()
```

```python
popular_df=num_rating_df.merge(avg_rating_df,on='Book-Title')
```

```python
popular_df
```

```python
popular_df=popular_df[popular_df['num_rating']>=250].sort_values('avg_rating',ascending=False).head(50)
```

```python
popular_df=popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_rating','avg_rating']]
```

```python
popular_df.shape
```

### collaborator based filtering system

```python
x=df_name.groupby('User-ID').count()['Book-Rating']>200
good_users=x[x].index
```

```python
filtered_rating=df_name[df_name['User-ID'].isin(good_users)]
```

```python
df_name
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books=y[y].index
```

```python
famous_books
```

```python
final_rating=filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
pt=final_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.shape
```

```python
pt.fillna(0,inplace=True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score=cosine_similarity(pt)
```

```python
similarity_score.shape
```

```python
def recommend(book_name):
    index=np.where(pt.index==book_name)[0][0]
    similar_items=sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]


    data=[]

    for i in similar_items:
        item=[]
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data
    
```

```python
recommend('Animal Farm')
```
